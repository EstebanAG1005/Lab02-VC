```python
from skimage.feature import hog
from skimage import data, exposure
import matplotlib.pyplot as plt
import cv2

# Cargar la imagen (ajusta la ruta de acuerdo a tu entorno local)
image_path = './img/person.png'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Asegurar que esté en formato RGB para la visualización

# Parámetros HOG similares a los que usaste en tus implementaciones manuales
orientations = 9
pixels_per_cell = (8, 8)
cells_per_block = (2, 2)

# Aplicar la función HOG de scikit-image
fd, hog_image = hog(image, orientations=orientations, pixels_per_cell=pixels_per_cell,
                    cells_per_block=cells_per_block, visualize=True, channel_axis=-1)

# Ajustar el contraste de la imagen HOG para una mejor visualización
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))

# Visualizar la imagen HOG
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(122)
plt.imshow(hog_image_rescaled, cmap='gray')
plt.title('Histograma de Gradientes Orientados')
plt.axis('off')

plt.show()
```
